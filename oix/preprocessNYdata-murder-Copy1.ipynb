```python
#other = affil.str.split(',')
#other.to_frame()
each = affil.str.split(';', expand=True)
each = each.ix[:, : 30]
each
```

```python
#get string with last author
lol = each.values.tolist()
lasts = []
for x in range(0,1999):
    #print "x %d" % x
    for y in range(0,30):
        #print "y %d" % y
        #print lol[x][y]
        if lol[x][y] is None:
            #print "null"
            continue
        else:
            strxy = ''.join(lol[x][y])
            if last3[x] in strxy:
                #print last3[x]
                #print strxy
                lasts.append(str(x) + "," + last3[x] + ",last," + str(citations[x]) +  "," + strxy)
#this contains only records with last authors
lasts
```
