```python
A = adjacency.values
```

```python
#https://stackoverflow.com/questions/29655111/igraph-graph-from-numpy-or-pandas-adjacency-matrix
# Create graph, A.astype(bool).tolist() or (A / A).tolist() can also be used.
g = igraph.Graph.Adjacency((A > 0).tolist())
```

```python
# Add edge weights and node labels.
g.es['weight'] = A[A.nonzero()]
g.vs['label'] = nodes  # or a.index/a.columns
gg = g.as_undirected()
```

```python
gg.es['weight'] = g.es['weight']
len(gg.es['weight'])
```

```python
#p, levels = gg.community_multilevel(return_levels=True) #produces 14 clusters
p, levels, clusts = gg.community_multilevel(weights='weight', return_levels=True)
print p
print levels
print clusts
q = gg.modularity(clusts)
print q
```

```python
for i, cluster in enumerate(clusts):
    print ("Cluster: " + str(i))
    print ("Size: " + str(len(cluster)))
    for item in cluster:
        print(nodes[item])
    print ("\n")
```

```python
nodes_clustered = pd.DataFrame(nodes)
#create second column for clusters
nodes_clustered['cluster'] = pd.Series(np.random.randn(len(nodes)), index=nodes_clustered.index)
#convert cluster column type from float to int
nodes_clustered['cluster'] = nodes_clustered['cluster'].astype(int)
#for each cluster, take each element and fill in cluster number into cluster column
for i, cluster in enumerate(clusts):
    for item in cluster:
        #print nodes[item] + "with index " + str(item) + " belongs to cluster " + str(i)
        nodes_clustered.cluster[item] = i
nodes_clustered
```

```python
strjson = '{"nodes":['
for node in nodes:
    #print node
    strjson = strjson + '{"name":"' + node + '","group":' + str(nodes_clustered.cluster[nodes.index(node)]) + '},'

#remove last comma from json
strjson = strjson[:-1]
print strjson
```
