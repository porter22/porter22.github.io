```python
#takes a filename with soup inside, returns a list of publications ids, with date, grantID, set type to "publications"
def getPublicationsFromFile(filename):
    publist = list()
    pubdict = dict()
    i = 0
    j = 0
    with open(filename) as file:  
        tempstr = file.read()
        soup = BeautifulSoup(tempstr, 'lxml')
        records = soup.find_all(re.compile("^oai:record"))
        print 'found ', len(records), ' records'
        for record in records:
            templist = list()
            acceptdate = record.find('dateofacceptance')
            if acceptdate.text == '':
                i = i + 1 #count records without a date
                continue #we do not add this record to the publication list
            title = record.find('title')
            code = record.find('code')
            pid = record.find('pid')
            if pid.text == '':
                #print '--------------------------'
                j = j + 1 
                continue #we do not add this record to the publication list
            #print pid.text, '-', acceptdate.text, '-', title.text, '-', code.text
            #templist.extend((title.text, code.text, acceptdate.text, 'publication'))
            publist.append([pid.text, acceptdate.text, title.text,code.text])
            #pubdict[pid.text] = templist
        print i, " records not included due to absence of date"
        print j, " records not included due to absence of doi"
        return publist;
```

```python
#make nodelist from grants and publications
def getNodelist(totalgrantlist, totalpublist):
    nodelist = list()
    for pub in totalpublist:
        nodelist.append([pub[0],pub[1],pub[2],pub[3], "publication"])
    for grant in totalgrantlist:
        nodelist.append([grant[0],grant[1], "grant"])
    return nodelist;
```

```python
#retrieves an edgelist in the form of grant - publication
def getEdgelist(totalgrantlist, totalpublist):
    edgelist = list()
    for index, grant in enumerate(totalgrantlist):
        #find publications that are connected to this grant
        print index
        for pub in totalpublist:
            #if pub.code matches grant.code
            #print pub[3], grant[0]
            if pub[3] == grant[0]:
                #print grant,' is related to: ', pub
                edgelist.append([grant, pub])
    return edgelist;
```

```python
#retrieves an edgelist where each publication has at least two connected grants
def getMultiEdgelist(totalgrantlist, totalpublist, connectedgrantsdict):
    edgelist = list()
    for pub in totalpublist:
        if pub[0] in connectedgrantsdict:
            #if there are several grants connected to this publication, we add each line to the edgelist
            if len(connectedgrantsdict[pub[0]]) > 1:
                for grant in connectedgrantsdict[pub[0]]:
                    print pub[0],"/", grant
                    edgelist.append([pub[0], grant])
    return edgelist;
```
